# BoxPlot

```python
import matplotlib
import pandas as pd
%matplotlib inline
```

```python
dataset = pd.read_csv('kc_house_data.csv')
matplotlib.style.use('ggplot')
dataset.boxplot(column='bedrooms')
```

```python
matplotlib.style.available
```

```python
dataset.boxplot(column='price', by='bedrooms')
```
