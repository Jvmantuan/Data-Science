# Consultando um dataset

- É possível realizar consultas em pandas como se fosse um SQL
- Pode-se usar operadores lógicos e instruções de agrupamento

```python
import pandas as pd
```

```python
file = 'kc_house_data.csv'
dataset = pd.read_csv(file)
dataset
```

O método `value_counts()` conta o número de valores únicos que uma coluna possui

```python
pd.value_counts(dataset['bedrooms'])
```

O método `loc` é usado para visualizar informações do dataset. Por exemplo, se quisermos localizar as casas com 3 quartos, podemos fazer o seguinte

```python
dataset.loc[dataset['bedrooms'] == 3]
```

Podemos usar o operador `&` para melhorar nossa busca.

Um exemplo: Queremos as casas que possuem 3 quartos e pelo menos 2 banheiros:

```python
dataset.loc[(dataset['bedrooms'] == 3) & (dataset['bathrooms'] > 2)]
```

Agora, vamos ordenar o dataset de acordo com o preço. Com o parâmetro `ascending=False`, dizemos que a ordenação será em ordem decrescente

```python
dataset.sort_values(by='price', ascending=False)
```

O pandas possui o método `count()` que conta o número de linhas do dataset. Note que ela pode ser usada para contar o número de linhas de uma busca, dessa forma:

```python
dataset[dataset['bedrooms'] == 4].count()
```

Como foi mostrado, existe 6881 casas com 4 quartos

## Alterando o dataset
Podemos criar novas colunas

```python
dataset['size'] = dataset['bedrooms'] * 2
```

```python
dataset.head()
```

Vamos criar uma função para processar os dados, e depois aplicar essa função em uma coluna nova

```python
# Criando a função 
def Categorizar(size):
    if size >= 8:
        return 'Big'
    elif size >= 6:
        return 'Medium'
    else:
        return 'Small'
```

```python
dataset['bedroom_category'] = dataset['size'].apply(Categorizar)
```

```python
dataset.head()
```

Vamos obter alguns dados desse dataset?

```python
pd.value_counts(dataset['bedroom_category'])
```

```python
dataset['bedroom_category'].mode()
```

Agora vamos deletar uma coluna usando o método `drop()`

```python
dataset.drop(['bedroom_category'], axis=1, inplace=True)
```

```python
dataset.drop(['size'], axis=1, inplace=True)
```

```python
dataset.head()
```

Podemos dropar linhas baseadas em uma condição lógica

```python
dataset.drop(dataset[dataset['bedrooms'] == 0].index, inplace=True)
```

```python
dataset.drop(dataset[dataset['bedrooms'] > 30].index, inplace=True)
```

```python
pd.value_counts(dataset['bedrooms'])
```
