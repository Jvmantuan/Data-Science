# Visualizando a distribuição de um dataset

## Histograma
- O método histograma procura o melhor número de bins

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
df = pd.DataFrame(tips)
```

```python
sns.set(color_codes=True)
```

```python
sns.distplot(tips.total_bill, kde=False)
```

**Plotando um histograma para um distribuição Normal**

```python
import numpy as np
x = np.random.normal(size=100)
```

```python
x
```

```python
sns.distplot(x)
```

**Plota um histograma com 20 bins e com KDE ativado**

```python
sns.distplot(tips.total_bill, bins=20, kde=True)
```

**Visualizando apenas o KDE**

```python
sns.distplot(tips.total_bill, hist=False)
```

**Parâmtro `rug`**

```python
sns.distplot(tips.total_bill, hist=False, rug=True)
```

## Jointplot (Scatter plot e Histograma)

- Usado quando queremos ver a distribuição de duas variáveis

```python
sns.jointplot(x='tip', y='total_bill', color='r', data=tips)
```

## Hexabin Plot

- Adequado para grandes bases de dados

```python
sns.jointplot(x='tip', y='total_bill', color='b', kind='hex', data=tips)
```

```python
sns.jointplot(x='tip', y='total_bill', color='b', kind='kde', data=tips)
```

## Visualizando relacionamentos emparelhados

- Plot de múltiplas variáveis de um dataset
- Cria uma matriz de eixos para mostrar a relação de cada par de colunas no dataframe
- Plota um histograma por padrão para cada coluna no dataframe na diagonal da matriz

```python
sns.pairplot(tips)
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', palette='Set1')
```

**Histograma com Regressão**

```python
sns.pairplot(iris, kind='reg')
```

**Plotando histogramas na diagonal**

```python
sns.pairplot(iris, hue='species', diag_kind='hist')
```

**Plotando apenas duas colunas do dataset**

```python
sns.pairplot(iris, hue='species', vars=["sepal_width", "sepal_length"], palette='husl')
```
