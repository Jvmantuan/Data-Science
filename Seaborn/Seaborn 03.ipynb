# BoxPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = tips = sns.load_dataset("tips")
```

**Boxplot de dias da semana por total da conta**

```python
sns.catplot(x="day", y='total_bill', kind='box', data=tips)
```

**Boxplot de dias da semana por total da conta e se a pessoa Ã© fumante**

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='box', data=tips)
```

# Boxen

- Tipo de grÃ¡fico focado na forma dos dados, nÃ£o nos outliers
- Interessante quando se tem grandes datasets

```python
sns.catplot(x="day", y='total_bill', kind='boxen', data=tips)
```

- O mÃ©todo `stripplot()` permite plotar a distribuiÃ§Ã£o dos dados
- Podemos combinar os dois grÃ¡ficos para ter mais informaÃ§Ãµes

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips)
sns.stripplot(x='day', y='total_bill', data=tips, color='gray')
```

**Usando o facetgrid com o catplot**

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='boxen', height=4)
```

# ViolinPlot

- Ideal para visualizar distribuiÃ§Ã£o de variÃ¡veis
- CombinaÃ§Ã£o do Boxplot com KDE
- Funciona para distribuiÃ§Ãµes com picos
- Permite uma visualizaÃ§Ã£o mais rica do que com o BoxPlot

**Violin Plot**

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

**Violin Plot Horizontal**

```python
sns.catplot(x='total_bill', y='day', kind='violin', data=tips)
```

**Usando o parÃ¢metro hue**

```python
sns.catplot(x='day', y='total_bill', hue='time', data=tips, kind='violin')
```

**Ã possÃ­vel dividir o grÃ¡fico usando o mÃ©todo `split`**

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, kind='violin', data=tips)
```
