# BoxPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
df = pd.DataFrame(tips)
```

```python
df
```

**Boxplot de dias da semana por total da conta**

```python
sns.catplot(x="day", y='total_bill', kind='box', data=tips)
```

**Boxplot de dias da semana por total da conta e se a pessoa é fumante**

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='box', data=tips)
```

# Boxen

- Tipo de gráfico focado na forma dos dados, não nos outliers
- Interessante quando se tem grandes datasets

```python
sns.catplot(x="day", y='total_bill', kind='boxen', data=tips)
```

- O método `stripplot()` permite plotar a distribuição dos dados
- Podemos combinar os dois gráficos para ter mais informações

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=tips)
sns.stripplot(x='day', y='total_bill', data=tips, color='gray')
```

**Usando o facetgrid com o catplot**

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='boxen', height=4)
```

# ViolinPlot

- Ideal para visualizar distribuição de variáveis
- Combinação do Boxplot com KDE
- Funciona para distribuições com picos
- Permite uma visualização mais rica do que com o BoxPlot

**Violin Plot**

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

**Violin Plot Horizontal**

```python
sns.catplot(x='total_bill', y='day', kind='violin', data=tips)
```

**Usando o parâmetro hue**

```python
sns.catplot(x='day', y='total_bill', hue='time', data=tips, kind='violin')
```

**É possível dividir o gráfico usando o método `split`**

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, kind='violin', data=tips)
```

- O parâmetro inner preenche o violin plot com observações
- Útil para destacar as partes que possuem mais dados

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, inner='stick', palette='pastel', kind='violin', data=tips)
```

**Combinando Swarm plot com Violin plot**

```python
sns.catplot(x='day', y='total_bill', hue='sex', split=True, palette='pastel', kind='violin', data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips)
```

```python
sns.catplot(x='day', y='total_bill', palette='pastel', kind='violin', data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips)
```
